```python
def func(beta, theta):
    if (5*beta*np.cos(theta) + beta*np.sin(theta) - 5) >= 0:
        return np.sin(theta)/2
    else:
        return 0
    
result = integrate.nquad(func, [[0,1],[0, np.pi]])
print(result)
```

```python
# Model 1: Gaussian
def likelihood1(x, mu, sig):
    return 1/(2*np.pi)**0.5/sig * np.exp(-(x-mu)**2/2/sig**2)

# Model 2: Student's t-distribution
from scipy.special import gamma as Ga

def likelihood2(x, mu, sig, v):
    return Ga((v+1)/2)/Ga(v/2)/sig/(np.pi*v)**0.5 * (1+(x-mu)**2/v/sig**2)**(-(v+1)/2)
```

```python
# draw the distributions
x = np.linspace(-5,5,101)
mu = 0
sig = 1
v = [1, 2, 5]

fig = plt.figure(figsize=(6,6))
plt.plot(x, likelihood1(x, mu, sig), label='Gaussian')
plt.plot(x, likelihood2(x, mu, sig, v[0]), label=r'$\nu = 1$')
plt.plot(x, likelihood2(x, mu, sig, v[1]), label=r'$\nu = 2$')
plt.plot(x, likelihood2(x, mu, sig, v[2]), label=r'$\nu = 5$')

plt.xlim(-5,5)
plt.legend()
plt.show()
```

```python
# Data Histogram
fig = plt.figure(figsize=(6,6))

# bins1 = np.linspace(-11,-7,9)
# bins2 = np.linspace(-11,-7,18)
hist_kwargs = {'bins':None, 'histtype':'step', 'density':True, 'stacked':True, 'linewidth':1.5}

plt.hist(data, **hist_kwargs)

plt.show()
```

```python
# Find the best parameters
def NLL1(p):
    mu, sig = p[0], p[1]
    return -np.sum(np.log(likelihood1(data, mu, sig)))

def NLL2(p):
    mu, sig, v = p[0], p[1], p[2]
    return -np.sum(np.log(likelihood2(data, mu, sig, v)))

opt1 = minimize(NLL1, x0=[data.mean(), 1.], tol=1e-8)['x']
opt2 = minimize(NLL2, x0=[data.mean(), 1., 2.], tol=1e-8)['x']

print(opt1, opt2)
```

```python
# BIC Calculation
BIC1 = 2*NLL1(opt1) + 2*np.log(len(data))
BIC2 = 2*NLL2(opt2) + 3*np.log(len(data))
dBIC = BIC1 - BIC2

print(BIC1, BIC2, dBIC)
```

```python
# Data Histogram
fig = plt.figure(figsize=(8,8))

# bins1 = np.linspace(-11,-7,9)
# bins2 = np.linspace(-11,-7,18)
hist_kwargs = {'bins':None, 'histtype':'step', 'density':True, 'stacked':True, 'linewidth':1.5}

plt.hist(data, **hist_kwargs, color='blue', label='Data')

x = np.linspace(-12, -1, 201)
plt.plot(x, likelihood1(x, *opt1), linestyle='--', color='red', label=rf'Model 1: $N({opt1[0]:.2f}, {opt1[1]:.2f}^2)$')
plt.plot(x, likelihood2(x, *opt2), linestyle='--', color='black', label=rf'Model 2: $t_{opt2[2]:.2f}({opt2[0]:.2f}, {opt2[1]:.2f}^2)$')

plt.xlabel(r'$M_V$')
plt.ylabel(r'$P(M_V|M, \theta)$')

plt.legend()
plt.savefig('midexam_5.pdf', dpi=300)
plt.show()
```
